## **1. What is Machine Learning?**

머신러닝이란, 컴퓨터가 데이터를 기반으로 학습해서 스스로 패턴을 찾아내고, 이를 통해 예측이나 결정을 내릴 수 있게 하는 기술이다. 예를 들어, 사진을 보여주면 어떤 동물인지 구분하거나, 미래의 주식 가격을 예측하는 등의 작업을 수행할 수 있다.   
쉽게 말하면, 머신러닝은 known data를 사용하여 unknown outcome을 예측하는 것이다.

## **2. Types of ML Problems**
머신 러닝은 label의 유무에 따라 크게 supervised learning과 unsupervised learning으로 나뉜다.

### Supervised Learning
지도학습(Supervised Learning)은 정답이 있는 데이터를 활용하여 데이터의 패턴과 모델을 알아내는 학습 방식을 말한다. 이때, 정답은 `Label`이라 부르고 주어진 데이터에 대해서 적잘한 Label을 예측하는 것을 목적으로한다. 지도학습의 데이터는 (Data + Label)의 쌍으로 이루어져 있다.

### **2-2. Unsupervised Learning (비지도학습)**

비지도학습은 label이 없는 데이터를 사용하여 데이터의 숨겨진 패턴이나 구조를 발견하는 알고리즘이다. 비지도학습의 종류는 매우 다양하지만, 대표적으로는 Clustering과 Anomaly Detection 등이 있다.

- Anomaly Detection (이상탐지) : 정상 패턴에서 벗어난 비정상적인 데이터를 식별하는 알고리즘. 예를 들어, 분당 70정도로 유지되던 심박수가 갑작스럽게 30으로 떨어지는 것을 비정상 심박동으로 감지하는 것을 anomaly detection이라고 한다.

### **3-1. Single-Variate Linear Regression**
- 정의 : 단변량 선형 회귀(Single-variate Linear Regression)은 주어진 데이터를 나타내는 최적의 `직선`을 찾아냄으로써 input (x)와 output (y) 사이의 관계를 도출해내는 과정이다.

- 공식 : $$\mathrm y=wx + b$$

  - $x$ : 독립변수 (입력변수)
  - $y$ : 종속변수 (출력변수)
  - $w$ : weight(가중치)
  - $b$ : bias(편향)   
  - weight는 기울기를, bias는 y절편에 해당된다. 학습은 이런 파라미터를 변경시키면서 값을 확인해 보는 과정을 말하는 것이다.

```python
# 필요 라이브러리 호출출
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# (1) load dataset (구글 드라이브 주소에 따라 변경)
df = pd.read_csv('/home/minho/Downloads/wk1/dataset/Salary_dataset.csv').set_index('Unnamed: 0')

# 데이터 일부 보기
print(df[:5])
```

```python
# (2) 데이터를 Training Set와 Test Set로 나눠주기
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df['YearsExperience'], df['Salary'], test_size=0.4, random_state=42)

# training set와 test set를 2차원 배열로 바꿔준다.
X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1, 1)
```

**가중치 계산**

가중치는 예측값과 실제값 사이의 오차를 구하고 최소화하는 방식으로 계산된다.

- 오차 구하는 방식

  - MAE(평균 절대 오차)    
  
    모델 예측값과 실제 라벨의 차이에 절댓값을 씌운 것의 평균(1차함수 형태이므로 최솟값을 찾기 힘들어 잘 사용하지 않음)

    -> 최소지점에서 미분이 불가능해, 이후 경사하강법을 이용한 최적화가 어렵다.  

    $MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$  


  - MSE(평균 제곱 오차)

    예측값과 실제 라벨의 차이에 제곱을 씌운 것의 평균,

    기울기가 0이 되는 지점(손실, 즉 오차가 최소)의 w를 가중치로 사용한다.

    -> MAE보다 Outlier(이상치)에 민감하다.


    $MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$  


  + RMSE(Root Mean Squared Error)

    MSE에 루트를 씌운 것

    -> MAE보다는 이상치에 민감하나, MSE에 비해서는 덜 민감하다


    $RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$

  - 2. RMSE가 MSE보다는 덜 민감하지만, MAE에 비해서는 이상치에 더 민감한 이유는 무엇일까요?

```python
# (3) LinearRegression 클래스 -> fit(), score(), predict()
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Linear Regression 모델을 훈련한다.
lr.fit(X_train, Y_train)

# YearsExperience : 5.0 에 대해 예측한다.
print(lr.predict([[5.0]]))
```

```python
# (4) 직선의 가중치(weight), 편향(bias) 출력하기
print(lr.coef_, lr.intercept_)
```

```python
# MAE 코드
from sklearn.metrics import mean_absolute_error
Y_pred = lr.predict(X_test)
print(Y_pred[:2], np.array(Y_test)[:2])
print(mean_absolute_error(Y_test, Y_pred))

# MSE 코드
from sklearn.metrics import mean_squared_error
Y_pred = lr.predict(X_test)
print(Y_pred[:2], np.array(Y_test)[:2])
print(mean_squared_error(Y_test, Y_pred))
```

# **Least Squares Solution 정리**

## **1. Least Squares Solution이란?**
- 선형회귀(Linear Regression)에서 **잔차(Residual)의 제곱합을 최소화하는 해**를 찾는 방법.
- 데이터가 주어졌을 때, **입력 $X$ 와 출력 $y$ 사이의 선형 관계를 가장 잘 설명하는 $w$ 를 구하는 방법**.

---

## **2. 선형회귀 모델**
선형회귀는 다음과 같은 형태를 가진다.

$$
y = Xw + \epsilon
$$

여기서,

- $y$ : 종속 변수(출력 값, Ground-Truth)
- $X$ : 독립 변수(입력 값)
- $w$ : 회귀 계수(가중치, 찾고 싶은 값)
- $\epsilon$ : 오차(Noise)

즉, 우리가 원하는 것은 **잔차(Residual) $y - Xw$ 의 크기를 최소화하는 $w$ 를 찾는 것**.

---

## **3. Least Squares Solution 유도**
최소제곱법은 **잔차의 제곱합을 최소화하는 문제**를 해결한다.

$$
J(w) = ||y - Xw||^2
$$

이를 **벡터 연산의 내적(inner product) 형태**로 전개하면,

$$
J(w) = (y - Xw)^T (y - Xw)
$$

이를 전개하면,

$$
J(w) = y^T y - 2 y^T Xw + w^T X^T X w
$$

여기서,
- $y^T y$ : **$w$ 와 무관한 상수 항**  
- $-2 y^T Xw$ : **$w$ 와 선형적으로 관계된 항**  
- $w^T X^T X w$ : **$w$ 에 대한 이차(quadratic) 항**

---

## **4. $w$ 에 대해 미분**
이제 $J(w)$ 를 **$w$ 에 대해 미분**하자.  
행렬 미분(Matrix Calculus) 공식을 이용하여 각각의 항을 따로 계산해 본다.

### **(1) 첫 번째 항: $y^T y$**
$$
\frac{d}{dw} y^T y = 0
$$
- $y^T y$ 는 **$w$ 와 무관한 상수**이므로 미분하면 **0**.

### **(2) 두 번째 항: $-2 y^T Xw$**
$$
\frac{d}{dw} (-2 y^T Xw) = -2 X^T y
$$
- $-2 y^T X$ 는 **상수 행렬**이고, $w$ 에 대해 선형(linear) 관계를 가짐.

### **(3) 세 번째 항: $w^T X^T X w$**
$$
\frac{d}{dw} (w^T X^T X w) = 2 X^T X w
$$
- 여기서 **$w^T A w$ 형태의 행렬 미분 공식**을 사용한다.

$$
\frac{d}{dw} (w^T A w) = 2 A w
$$

- 여기서 $A = X^T X$ 이므로,

$$
\frac{d}{dw} (w^T X^T X w) = 2 X^T X w
$$

---

## **5. 미분 결과 정리**
각 항을 종합하여, **손실 함수 $J(w)$ 의 미분 결과는**:

$$
\nabla_w J(w) = 2X^T X w - 2X^T y
$$

이제 최소점을 찾기 위해 **미분값을 0으로 설정**하면,

$$
2X^T X w = 2X^T y
$$

양변을 2로 나누면,

$$
X^T X w = X^T y
$$

이것이 바로 **정규방정식(Normal Equation)** 이다.

---

## **6. Least Squares Solution 도출**
위의 방정식을 풀기 위해 **역행렬을 양변에 곱하면**, 최적의 $w$ 를 얻을 수 있다.

$$
w^* = (X^T X)^{-1} X^T y
$$

이것이 바로 **Least Squares Solution(최소제곱 해)** 이다.

---

## **7. Least Squares Solution의 한계점**
1. **다중공선성(Multicollinearity) 문제**  
   - $X^T X$ 의 역행렬이 존재하지 않을 수도 있음.  
   - 해결책: **정규화 회귀(Ridge, Lasso)** 사용.

2. **고차원 데이터에서 비효율적**  
   - $(X^T X)^{-1}$ 계산의 시간 복잡도가 $O(p^3)$ 이므로, 특성 수 $p$ 가 클 경우 비효율적.
   - 해결책: **QR 분해, SVD, 경사하강법 사용**.

3. **대규모 데이터셋에서 처리 어려움**  
   - 역행렬 계산이 부담되므로 **경사하강법(Gradient Descent)** 을 사용하여 최적해를 점진적으로 찾는 것이 효과적.

---

## **8. Least Squares Solution을 구하는 다양한 방법**
| 방법 | 시간 복잡도 ($O$) | 특징 |
|------|----------------|------|
| **정규방정식 (Normal Equation)** $w^* = (X^T X)^{-1} X^T y$ | $O(p^3)$ | 해석적(Closed-form) 해를 한 번에 구할 수 있지만, 역행렬 계산이 부담됨. |
| **LU 분해 (LU Decomposition)** | $O(p^3)$ | 역행렬을 직접 구하지 않고 선형 방정식을 풀이. |
| **QR 분해 (QR Decomposition)** | $O(n p^2)$ | 큰 데이터셋에서 $X$ 의 열이 많을 때 안정적인 방법. |
| **SVD 분해 (Singular Value Decomposition)** | $O(n p^2)$ | 다중공선성 문제를 해결할 수 있음. |
| **경사하강법 (Gradient Descent)** | $O(n p)$ (반복 횟수에 따라 변동) | 대규모 데이터셋에서 사용, 점진적으로 최적해를 찾음. |

---

## **9. 언제 어떤 방법을 사용할까?**
| 데이터 크기 & 특성 수 | 추천 방법 |
|-----------------|------------|
| 작은 데이터셋 ($p$ 작음) | 정규방정식 $(X^T X)^{-1} X^T y$ |
| 다중공선성 문제 있음 | SVD 분해 |
| 데이터 차원 $p$ 이 크다 | QR 분해 |
| 데이터가 매우 많음 ($n$ 큼) | 경사하강법 |

- **작은 데이터셋이면 정규방정식 사용**
- **차원이 클 경우 QR 분해 또는 SVD 사용**
- **대규모 데이터셋이면 경사하강법 사용**

---

## **10. 최종 정리**
- **Least Squares Solution은 선형회귀에서 잔차의 제곱합을 최소화하는 해를 찾는 방법**.
- **해석적 해(Closed-form Solution)로 $w^* = (X^T X)^{-1} X^T y$ 를 사용하지만, 역행렬 계산이 부담될 수 있음**.
- **데이터 크기에 따라 LU 분해, QR 분해, SVD, 경사하강법 같은 다양한 최적화 방법을 사용 가능**.
- **실제 적용할 때는 데이터의 크기와 특성 수에 따라 적절한 방법을 선택해야 함**.

즉, **Least Squares Solution은 잔차의 제곱합을 최소화하는 최적의 $w$ 를 찾는 과정이며, 행렬 미분을 통해 유도된다!** 🚀


```python
# 시각화
plt.scatter(X_train, Y_train)
plt.plot([1,10], [1*lr.coef_+lr.intercept_, 10*lr.coef_+lr.intercept_])

plt.scatter(5.0, 72303.5, marker='^')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()

# 이제는 Training set 범위에서 벗어난 YearsExperience에 대해서도 Salary를 예측할 수 있게 되는 것이다.
```

### **3-2. Polynomial Regression**


아래의 그래프와 같이 데이터의 분포가 직선 형태가 아닐 때는 어떻게 해야할까? 이때 다항회귀를 사용한다.


```python
# 이런 그래프는 일반적인 직선으로 해결하기 어려움.
n = 100

x = 6 * np.random.rand(n, 1) - 3
y = 0.5 * x**2 + x + 2 + np.random.rand(n, 1)

plt.scatter(x, y, s=5)
```

- 정의 : 다항 회귀(Polynomial Regression)는 주어진 데이터를 나타내는 최적의 `다항식`을 찾아냄으로써 input (x)와 output (y) 사이의 관계를 도출해내는 과정이다.

- 공식 : $$y=w_nx^n+w_{n-1}x^{n-1}+...+w_2x^2+w_1x+w_0$$

  - $x$ : 독립변수 (입력변수)
  - $y$ : 종속변수 (출력변수)
  - $W=(w_0,w_1,...,w_n)$ : parameters ($w_0$ = bias)
  
  - 이때, $w_0$은 단변량 선형회귀에서의 bias term을 일반화한 표현이다.
  - scikit-learn에서 PolynomialFeatures를 사용하면 된다. 


```python
from sklearn.preprocessing import PolynomialFeatures
# x를 x^2로 바꿔서 x^2, y로 보겠다.
poly = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly.fit_transform(x)
print(poly.get_feature_names_out())
print(x[0])
print(x_poly[0])
```

```python
lr_1 = LinearRegression()
lr_1.fit(x_poly, y)

# 다항회귀 그래프
xx = np.linspace(-3, 3, 200)
xx_poly = poly.transform(xx.reshape(-1, 1))
y_pred = lr_1.predict(xx_poly)

plt.scatter(x, y, s=3)
plt.plot(xx, y_pred, 'r', label = 'prediction')
plt.legend()
plt.show()
```

- 공식 : $$y=w_1x_1+w_2x_2+...+w_nx_n+w_0$$

  - $y$ : 종속변수 (출력변수)

$y= w_3x_3+w_2x_2+w_1x_1+w_0$

### **4-1. Overfitting and Underfitting**

도현, 수민, 민수가 수능을 준비하고 있다고 생각해보자.   
도현이는 어쩌다보니 모의고사 답지를 구하게 되어서, 공부를 하지 않고 그 답지만 전부 외워서 모의고사에 응시했다. 도현이는 모의고사를 전부 만점을 맞고 의기양양했지만, 결국 수능은 전과목 9등급을 받고 말았다.   
그에 반해, 수민이는 정정당당하게 공부를 했다. 비록 도현이보다 모의고사 점수는 낮았지만, 공부를 열심히 했기 때문에 수능은 도현이보다 훨씬 잘 봤다.   
민수는 공부에 별로 열의가 없었기 때문에 개념 공부도 제대로 하지 않고 모의고사와 수능을 응시했다. 결국 민수는 모의고사 점수와 수능 점수 전부 형편없었다.

Machine learning에서도 동일하다.   
모델이 도현이처럼 train set(모의고사)에만 과도하게 적합하면 train 성능은 좋지만 test set(수능)에 대한 성능이 떨어질 수 있다. 민수처럼 모델이 train set도 제대로 학습하지 않는다면 train 성능과 test 성능 모두 좋지 않을 것이다. 따라서 수민이와 같은 적정한 수준의 학습이 필요하다.   
이때, 도현이는 overfit(과대적합)된 상태, 수민이는 optimal한 상태, 민수는 underfit(과소적합)된 상태라고 할 수 있다.

- overfit (과대적합): 모델이 학습데이터에 너무 과하게 맞춰져, 새로운 데이터(test data)에서 성능이 떨어지는 현상. 학습 데이터의 노이즈까지 학습하는 경우 발생한다.
- underfit(과소적합) : 모델이 학습데이터조차 제대로 학습하지 못한 상태. 모델이 너무 단순하거나 학습이 충분하지 않은 경우 발생한다.

### **4-2. Regularization**

정규화(규제, Regularization)란 모델이 학습데이터에 과대적합되는 것을 방지하기 위해 모델의 복잡성을 제어하는 기법이다. 모델이 학습 데이터에 지나치게 적합되지 않도록 penalty를 추가하여 일반화 성능(새로운 데이터에 대한 성능)을 향상시키는 것이 목표이다.  
따라서, 아래의 수식과 같이 모델의 loss에 penalty term이 추가된다. 이때 $\lambda$는 hyperparameter로써, regularization의 강도를 조절한다.
Linear regression model에 regularization을 추가한 모델은 크게 두 가지, lasso와 ridge가 있다.  

1) **Ridge Regression**
- L2 regularization 이용 (L2 norm을 이용한 regularization) : 원래 사용하던 MSE에 '가중치 제곱'의 추가 규제를 줘서 계수를 줄인다.
- 파라미터의 크기가 큰 것을 빠르게 줄여준다.
-  `sklearn.linear_model`에서 `Ridge`를 import해 사용 가능하다.
- Loss 함수에 가중치의 제곱을 추가해서, 계수(가중치)의 합이 커지지 않도록 규제하는 것

$$\mathrm J(\theta) = MSE(\theta) + \lambda\sum_{j=0}^{p}\theta{j}^2$$

2) **Lasso Regression**
- L1 regularization (L1 norm을 이용한 regularization) : 원래 사용하던 MSE에 '가중치 절댓값'의 추가 규제를 줘서 계수를 줄인다.
- Ridge와는 달리, regularization term이 제곱이 아닌 절댓값이므로, 파라미터의 크기와 상관없이 같은 수준의 규제가 적용된다.
- feature selection 효과가 있다.
- `sklearn.linear_model`에서 `Lasso`를 import해 사용 가능하다.
$$\mathrm J(\theta) = MSE(\theta) + \lambda\sum_{j=0}^{p}|\theta{j}|$$

## 선형 회귀에서 "계수가 크다"는 의미

### 1. 계수가 크다는 것의 의미  
   예측값은 다음과 같이 나타낼 수 있습니다:  
   $$
   y = w_1 x_1 + w_2 x_2 + \dots + w_n x_n + b
   $$
   - $w_i$가 클수록 해당 $x_i$의 변화에 따라 예측값 $y$가 크게 변합니다.
   - 작은 변화에도 민감하게 예측이 바뀌므로 노이즈에 과적합할 가능성이 큽니다.
   - 즉, 특정 변수에 대한 의존도가 지나치게 커질 수 있으며, 이는 일반화 성능을 떨어뜨릴 수 있습니다.

### 2. 계수를 줄이면 왜 과적합이 줄어드는가?  
   Ridge 회귀는 큰 계수에 대해 벌점을 부과하여(아래와 같은 $\ell_2$ 노름 사용) 계수를 작게 만듭니다.
   - 계수가 크면 특정 변수에 대한 의존이 과도해져 과적합 위험이 증가합니다.
   - 계수를 작게 유지하면 여러 변수에 적절히 분산해 의존하므로 노이즈에 과민하게 반응하는 것을 줄일 수 있습니다.
   - 결과적으로 새로운 데이터가 들어왔을 때 더 일반화가 잘 되는 효과가 있습니다.

### 3. Ridge는 왜 계수를 완전히 0으로 만들지 못할까?  
   Ridge에서는 다음의 제곱 페널티를 사용합니다:
   $$
   \lambda (w_1^2 + w_2^2 + \dots + w_n^2)
   $$
   - $\ell_2$ 노름은 $w = 0$ 근처에서 평평한 형태를 띠어, 계수를 정확히 0으로 만드는 것보다 작게 유지하는 것이 오히려 페널티 감소 효과가 낫습니다.
   - 반면에 Lasso $\ell_1$ 노름는 $|w|$ 형태로, 0 근처에서 기울기가 일정하게 유지되므로 계수를 0으로 만들 가능성이 큽니다.
   - 따라서 Ridge는 계수를 줄이되, 꼭 0으로 만들 필요가 없다고 판단하기 때문에 0이 아닌 작은 값으로 남아 있게 됩니다.


```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
import numpy as np

housing = fetch_california_housing()

X_train, X_test, y_train, y_test = train_test_split(housing['data'], housing['target'], test_size = 0.4, random_state=0)

lasso = Lasso(alpha=0.3).fit(X_train, y_train) #Ridge로도 변경가능

print("Training Set 점수 : {:.2f}".format(lasso.score(X_train, y_train)))
print("Test Set 점수 : {:.2f}".format(lasso.score(X_test, y_test)))
```
